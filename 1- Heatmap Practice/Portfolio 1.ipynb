```python
response = requests.get('https://api.sofascore.com/api/v1/event/11451079/player/226986/heatmap', headers=headers)
#İrfan Can
heatmap = response.json()
heatmap
```

```python
# Plotting

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches
from mplsoccer.pitch import Pitch, VerticalPitch
```

```python
df_heatmap = pd.DataFrame(heatmap['heatmap'])
df_heatmap
```

```python
df_heatmap['x'] = df_heatmap['x']*1.2
df_heatmap['y'] = df_heatmap['y']*.8
```

```python
pitch = VerticalPitch(positional=True, shade_middle=True, positional_color='#eadddd', shade_color='#f2f2f2')
fig, ax = pitch.draw(figsize=(13, 8))

# Create a heatmap using Seaborn
sns.scatterplot(x='y', y='x', data=df_heatmap, ax=ax, marker='o', s=10, color='blue')
sns.kdeplot(data=df_heatmap, x='y', y='x', cmap='Reds', shade=True,shade_lowest=False,alpha=.25,n_levels=10,)

# Calculate the average coordinates
average_x = df_heatmap['y'].mean()
average_y = df_heatmap['x'].mean()

# Add a large point and label "17" inside the dot
ax.scatter(average_x, average_y, s=400, color='blue', edgecolor='black', marker='o')
ax.annotate('17', (average_x, average_y), color='yellow', fontsize=12, ha='center', va='center')

# Show the plot
plt.title("İrfan Can Kahveci vs Pendikspor")
plt.xlim(0,80)
plt.ylim(0,120)
plt.gca().invert_xaxis()
plt.savefig('irfan_vs_pendikspor.png', bbox_inches='tight')
plt.show()
```
