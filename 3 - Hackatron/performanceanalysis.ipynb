```python
file_path = 'Premier_League_team_data_CSV.csv'
df = pd.read_csv(file_path, sep=';')
df
```

```python
# İstenmeyen sütunları belirleme
not_avg = ['team_name', 'season_name', 'competition_name', 'team_name2','current_league'] # average, 2023/2024, Premier League (ENG), average, PL

avgs = df.drop(columns=not_avg).mean()

liverpool = df[df['team_name'] == 'Liverpool FC']

# avgs DataFrame'ine sütunlar ekleyerek genişletme
avgs['team_name'] = 'average'
avgs['season_name'] = '2023/2024'
avgs['competition_name'] = 'Premier League (ENG)'
avgs['team_name2'] = 'average'
avgs['current_league'] = 'PL'

avgs
```

```python
top6 = df[df['team_name'].isin(['Liverpool FC', 'Arsenal FC','Manchester United FC','Manchester City FC','Tottenham Hotspur FC','Aston Villa FC'])]
top6_avgs = top6.drop(columns=not_avg).mean()

top6_avgs['team_name'] = 'top6average'
top6_avgs['season_name'] = '2023/2024'
top6_avgs['competition_name'] = 'Premier League (ENG)'
top6_avgs['team_name2'] = 'top6average'
top6_avgs['current_league'] = 'PL'
top6_avgs
```

```python
# Set seaborn theme
sns.set_theme(style="whitegrid")

# Make the PairGrid
g = sns.PairGrid(selected_data.sort_values("team_name", ascending=False),
                 x_vars=selected_data.columns[1:], y_vars=["team_name"],
                 height=10, aspect=.25)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=10, orient="h", jitter=False,
      palette="flare_r", linewidth=1, edgecolor="w")

# Use semantically meaningful titles for the columns
titles = selected_data.columns[1:]

for ax, title in zip(g.axes.flat, titles):
    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

# Set x-axis limit and label
g.set(xlim=(0, 100), xlabel="Value", ylabel="")

plt.show()
```

```python
# Takım adlarını dönüştürmek için bir sözlük oluştur
team_name_mapping = {
    'avgs': 'Average',
    'liverpool': 'Liverpool',
    'top6_avgs': 'Top 6 Average'
}

# Özellik etiketlerini dönüştürmek için bir sözlük oluştur
selected_labels_mapping = {
    'offensiveDuels_eff': 'Offensive Duels Efficiency',
    'passes_accuracy': 'Passes Accuracy',
    'longPasses_accuracy': 'Long Passes Accuracy',
    'shots_on_target_rate': 'Shots on Target Rate',
    'progressive_passes_rate': 'Progressive Passes Rate',
    'aerial_duels_in_final_3rd_efficiency': 'Aerial Duels in Final Third Efficiency',
    'pass_accuracy_own_half_rate': 'Pass Accuracy in Own Half Rate',
    'pass_accuracy_opponent_half_rate': 'Pass Accuracy in Opponent Half Rate',
    'off_aerial_duels_eff': 'Offensive Aerial Duels Efficiency',
    'forward_pass_accuracy_rate': 'Forward Pass Accuracy Rate'
}

# Seçili özellikler
selected_labels = ['offensiveDuels_eff',
                   'passes_accuracy',
                   'longPasses_accuracy',
                   'shots_on_target_rate',
                   'progressive_passes_rate',
                   'aerial_duels_in_final_3rd_efficiency',
                   'pass_accuracy_own_half_rate',
                   'pass_accuracy_opponent_half_rate',
                   'off_aerial_duels_eff',
                   'forward_pass_accuracy_rate'
                   ]

# Etiketler (özellikler) ve takımlar
labels = [selected_labels_mapping[label] for label in selected_labels]
teams = selected_data['team_name']

# Number of variables we're plotting.
num_vars = len(labels)

# Split the circle into even parts and save the angles
# so we know where to put each axis.
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to "complete the loop"
# and append the start value to the end.
angles += angles[:1]

def plot_radar(data, ax, color):
    # Veriyi tamamlayın
    values = data[selected_labels].values.tolist()
    values += values[:1]  # Döngüyü tamamla

    # Liverpool takımının rengini belirle
    if data['team_name'] == 'Liverpool':
        color = '#C8102E'  # Liverpool takımı için belirlenen renk kodu

    # Radar grafiğini çiz
    ax.plot(angles, values, color=color, linewidth=1, label=data['team_name'])
    ax.fill(angles, values, color=color, alpha=0.25)


# fig ve ax oluştur
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Her takım için radar grafiğini çiz
for i, team in enumerate(teams):
    # Takım adını dönüştür
    team_name = team_name_mapping.get(team, team)
    # Güncellenmiş takım adını veri setindeki takım adıyla değiştir
    selected_data.at[i, 'team_name'] = team_name
    # Radar grafiğini çiz
    plot_radar(selected_data.loc[i], ax, color=plt.cm.viridis(i / len(teams)))

# Eksenleri düzelt
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Her açı için etiketleri ekle
ax.set_thetagrids(np.linspace(0, 360, num_vars, endpoint=False), labels)

# Satır adlarını düzenle
ax.set_xticklabels(labels, fontsize=16)  # Yazıların büyüklüğünü iki katına çıkar

# Efsaneyi güncelle
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=12)  # Yazıların büyüklüğünü iki katına çıkar

# Etiketlerin hizalamasını ayarla
for label, angle in zip(ax.get_xticklabels(), angles):
    if angle in (0, np.pi):
        label.set_horizontalalignment('center')
    elif 0 < angle < np.pi:
        label.set_horizontalalignment('left')
    else:
        label.set_horizontalalignment('right')

# Radar grafiği sınırlarını ayarla
ax.set_ylim(0.3, 1)

# Etiket pozisyonunu ayarla
ax.set_rlabel_position(180 / num_vars)

# Bazı özel stil ayarları
ax.tick_params(colors='#222222', labelsize=12)  # Yazıların büyüklüğünü iki katına çıkar
ax.grid(color='#AAAAAA')
ax.spines['polar'].set_color('#222222')
ax.set_facecolor('#FAFAFA')

# Başlık ekle
ax.set_title('Comparing Teams Across Metrics', y=1.08, fontsize=16)  # Yazıların büyüklüğünü iki katına çıkar

plt.show()
```
