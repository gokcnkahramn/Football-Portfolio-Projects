```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt

# Verileri tanmla
years = ['2019-20', '2020-21', '2021-22', '2022-23']
defensive_third = [5.13, 5.13, 5.45, 6]
middle_third = [6.39, 6.08, 6.32, 6.58]
attacking_third = [2.95, 2.63, 2.82, 2.95]
interceptions = [9.26, 9.13, 8.95, 8.76]
mistakes = [0.66, 0.53, 0.21, 0.53]

# Bar grafiini olutur
plt.figure(figsize=(12, 6))

# Bar chart'lar iz
bar_width = 0.15
index = range(len(years))
colors = ['blue', 'green', 'orange', 'red', 'purple']
labels = ['Tackles in Defensive Third', 'Tackles in Middle Third', 'Tackles in Attacking Third', 'Interceptions', 'Mistakes Leads Shot']

for i, data in enumerate(zip([defensive_third, middle_third, attacking_third, interceptions, mistakes], colors, labels)):
    plt.bar([idx + i * bar_width - 0.3 for idx in index], data[0], color=data[1], width=bar_width, label=data[2])
    for idx, val in zip(index, data[0]):
        plt.annotate(str(val), xy=(idx + i * bar_width - 0.3, val), xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

# Line graph'lar iz
for i, data in enumerate(zip([defensive_third, middle_third, attacking_third, interceptions, mistakes], colors)):
    plt.plot([idx + i * bar_width - 0.3 for idx in index], data[0], marker='o', linestyle='-', color=data[1])

# Balk ve etiketler ekle
plt.title('Liverpool Defensive Actions by Year per 90 min')
plt.xlabel('Years')
plt.ylabel('Values')
plt.xticks(index, years)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Grsel gelitirmeler
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
